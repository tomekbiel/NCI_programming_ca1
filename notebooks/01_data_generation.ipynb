```python
# Import required libraries
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Add src directory to path
sys.path.append(os.path.abspath(os.path.join('..', 'src')))

from data_generator import StudentGenerator
```
